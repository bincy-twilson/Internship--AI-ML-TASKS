```python

import pandas as pd

dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset=pd.read_csv('Social_Network_Ads.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
print(y_test)
print(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
clf2=DecisionTreeClassifier(
    max_depth=3,
    min_samples_split=10,
    min_samples_leaf=5,
    random_state=42
)
clf2.fit(x_train,y_train)
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
print(clf2.predict(sc.transform([[51,87000]])))
```

```python
y_pred=classifier.predict(x_test)
y_pred2=clf2.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
ac=accuracy_score(y_test,y_pred)    
print(cm)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm2=confusion_matrix(y_test,y_pred2)
ac2=accuracy_score(y_test,y_pred2)
print(cm2)
print(ac2)
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt

DecisionBoundaryDisplay.from_estimator(
    classifier,
    x_train,
    response_method='predict',
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1],
)
plt.scatter(x_train[:,0],x_train[:,1],c=y_train,s=20,edgecolors='k')
plt.title('Decision Tree Classifier (Training set)')
plt.show()
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
import matplotlib.pyplot as plt
DecisionBoundaryDisplay.from_estimator(
    classifier,
    x_test,
    response_method='predict',
    cmap=plt.cm.Spectral,
    alpha=0.8,
    xlabel=dataset.columns[0],
    ylabel=dataset.columns[1],
)
plt.scatter(x_test[:,0],x_test[:,1],c=y_test,s=20,edgecolors='k')
plt.title('Decision Tree Classifier (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.show()  
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
def plot_decision_tree(clf_object, feature_names,class_names):
    plt.figure(figsize=(15, 10))
    tree.plot_tree(
        clf_object,
        filled=True,
         feature_names=feature_names,
         class_names=class_names,
         rounded=True,
         )
    plt.show()
plot_decision_tree(classifier,['Age','Estimated Salary'],['Not Purchased','Purchased'])
```

```python
importances=classifier.feature_importances_
print("FEATURE IMPORTANCES:")
for name, importance in zip(['Age','Estimated Salary'], importances):
    print(f"{name}: {importance:.4f}")
```

```python
import pandas as pd
import numpy as np  
from sklearn.datasets import load_iris
from sklearn import tree
import graphviz

iris = load_iris()
df_iris=pd.DataFrame(iris['data'], columns=iris['feature_names'])
df_iris['target'] = iris['target']
df_iris.head()
clf = tree.DecisionTreeClassifier()
clf=clf.fit(iris.data, iris.target)
dot_data = tree.export_graphviz(clf, out_file=None, 
                                feature_names=iris.feature_names,  
                                class_names=iris.target_names,  
                                filled=True, rounded=True,  
                                special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
